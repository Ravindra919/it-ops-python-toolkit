```python
#The long way — using a loop:
cpus = [95, 45, 88, 30, 72]
high_cpu = []
for cpu in cpus:
    if cpu > 80:
        high_cpu.append(cpu)
print(high_cpu)
```

```python
#The same above using List Comprehension
cpus = [95,45,88,30,72]
high_cpu = [cpu for cpu in cpus if cpu > 80]
print(high_cpu)
```

```python
numbers = [1,2,4,5,6,7,8]
squared = [n*n for n in numbers]
print(squared)
```

```python
disks = [95,40,88,30,72]
alerts = [d for d in disks if d > 90]
print(alerts)
```

```python
#Add lable to each server
servers = ["SERVER01", "SERVER02", "SERVER03"]
labeled = [f"HOST: {s}" for s in servers]
print(labeled)
```
