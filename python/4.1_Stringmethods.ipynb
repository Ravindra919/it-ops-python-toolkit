```python


PART 5 — find() and count()
pythonlog = "ERROR: connection failed. ERROR: timeout"

# find() — returns position of first match, -1 if not found
print(log.find("ERROR"))        # 0 — found at position 0
print(log.find("WARNING"))      # -1 — not found

# count() — how many times does it appear
print(log.count("ERROR"))       # 2

PART 6 — Removing % and converting to int
You will need this a lot when parsing log values:
pythonvalue = "91%"

# Step 1 — remove the % sign
cleaned = value.replace("%", "")
print(cleaned)      # 91  — but still a string!

# Step 2 — convert to integer
number = int(cleaned)
print(number)       # 91  — now a number you can compare

# One line version
number = int(value.replace("%", ""))
print(number + 9)   # 100 — now you can do math on it

PART 7 — Case insensitive check
pythonlog = "WARNING: High Memory Usage on DB-Server-01"

# Problem — this fails
print("memory" in log)              # False — capital M

# Solution — convert both to lowercase first
print("memory" in log.lower())      # True
Always use .lower() when searching — log formats are inconsistent in real environments.

PART 8 — Putting it all together
pythonlog_line = "2026-03-04 | SERVER01 | CPU | 91%"

# Split into parts
parts = log_line.split(" | ")

# Extract each field
date   = parts[0]
server = parts[1]
metric = parts[2]
value  = int(parts[3].replace("%", ""))

# Apply logic
if value > 85:
    status = "CRITICAL"
elif value > 70:
    status = "WARNING"
else:
    status = "OK"

print(server, "|", metric, "|", str(value) + "%", "|", status)
# SERVER01 | CPU | 91% | CRITICAL

Quick Reference Card
MethodWhat it doesExample.strip()Remove whitespace"  hi  ".strip() → "hi".upper()All caps"hello".upper() → "HELLO".lower()All small"HELLO".lower() → "hello".replace(a,b)Replace a with b"ERROR".replace("ERROR","ALERT").split(x)Break into list"a,b".split(",") → ["a","b"].find(x)Position of x"ERROR".find("R") → 3.count(x)Count of x"aa".count("a") → 2x in stringCheck if exists"E" in "ERROR" → True.startswith(x)Starts with x"ERROR".startswith("E") → True.endswith(x)Ends with x"ERROR".endswith("R") → True

Now you have everything you need. Try Exercise 1 in Jupyter and paste your code! 🚀
```
