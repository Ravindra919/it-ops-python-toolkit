```python
import numpy as np

# Each row is a student, each column is a subject
marks = np.array([
    [45, 78, 90],   # shiva
    [58, 82, 76],   # john
    [92, 88, 95],   # marry
    [98, 74, 88]    # ram
])

# 1 - print shape
print("Shape:", marks.shape)  

# 2 - average score of each student (row wise)
print("Student averages:", marks.mean(axis=1))  

# 3 - average score of each subject (column wise)
print("Subject averages:", marks.mean(axis=0))   
```

```python
# Dataset — 5 students, 3 subjects (Math, Science, English)
import numpy as np
marks = np.array([
    [45, 78, 90],
    [58, 82, 76],
    [92, 88, 95],
    [98, 74, 88],
    [40, 55, 60]
])

print("shape", marks.shape)
print("average of each student", marks.mean(axis = 1))
print("average of each subject", marks.mean(axis = 0))
print("highest scored in the entire database", marks.max())
print("gives true or false for the conditions student average > 80", marks.mean(axis=1)>80) #boolean mask
print("students scored above 80 average", marks[marks.mean(axis=1)>80])  #boolean masking for dataframe, only prints where condition is true.
print("multiply", marks*1.1)
```

```python
#Dataset - 5 servers, 6 time steps for cpu_metrics
import numpy as np
import numpy as np
cpu_metrics = np.array([
    [84, 87, 90, 83, 94, 150],   
    [78, 57, 78, 56, 89, 91],
    [56, 74, 46, 58, 69, 79],
    [87, 76, 65, 54, 94, 160],   
    [60, 62, 61, 59, 63, 64]
])
print(cpu_metrics)
print("shape", np.shape(cpu_metrics))
print("shape", cpu_metrics.shape)   #cannot use .shape()...gives error...it works in 1D only becoz shape is an attribute...only methods have ()
print("average of all server cpu readings-global", cpu_metrics.mean)
print("standard deviation of all servers-global", cpu_metrics.std)
print("average of each server", cpu_metrics.mean(axis=1))  #server wise
print("average of all servers at particular time ", cpu_metrics.mean(axis=0))  #time wise for all servers
print("high cpu among all servers", cpu_metrics.max())
print("servers with above 80 average", cpu_metrics[cpu_metrics.mean(axis=1) >80]) 
print("multiply server metrics with 1.1", cpu_metrics * 1.1)
```
